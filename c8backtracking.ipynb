class Solution(object):
    def permuteUnique(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        
        result = []
        nums.sort()
        isUsed =[False] * len(nums)

        def backtracking(arr, perm):

            if len(perm) == len(nums):
                result.append(perm[:])
                return
            

            for i in range(len(arr)):
                if i > 0 and arr[i] == arr[i-1] and isUsed[i-1] == False:
                    # we dont wanna use same number to start a new branch
                    continue

                # else they are not same number
                if(not isUsed[i]):
                    isUsed[i] = True
                    perm.append(arr[i])
                    backtracking(arr, perm)
                    perm.pop()
                    isUsed[i] = False
            
        backtracking(nums, [])
        return result
