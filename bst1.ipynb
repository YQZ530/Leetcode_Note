        if not node:
            return

def preorderStact(root):
    if not root:
        return
    
    stack = [root]
    result = []

    while stack: # stack is not empty
        node = stack.pop()
        result.append(node.val)
        # 先压右，再压左（因为栈是后进先出）
        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)

    return result

class Solution(object):
    def rightSideView(self, root):
        if not root:
            return []

        queue = deque([root])
        result = []

        while queue:
            size = len(queue)
            curlvl =[]
            for i in range(size):
                node = queue.popleft()

              
                curlvl.append(node.val)

                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            result.append(curlvl)
        return result
