```python



class Solution(object):
    def sumNumbers(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """

        def dfs(node, sumv) -> int:
            if not node:
                return 0

            sumv = sumv *10 + node.val
            left,right =0,0

            if not node.left and not node.right: # very important when node is leaf node,
                return sumv

            if node.left:
                left = dfs(node.left, sumv)
            if node.right:
                right = dfs(node.right, sumv)
            
            return left + right
        return dfs(root, 0)
```
